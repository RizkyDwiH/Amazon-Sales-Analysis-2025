# Amazon Sales Analysis 2025

##  Exploratory data analysis

This project provides an in-depth analysis of Amazon's sales data for the year 2025. The goal is to extract actionable business insights through Python-based data exploration and visualization.  

Data preprocessing and cleaning were performed using SQL to ensure accuracy and reduce noise. The final analysis is executed in Python using the `pandas`, `matplotlib`, and `seaborn` libraries.  

The notebook covers the following key areas:

- **Total sales by product category**  
- **Monthly sales trends**  
- **Top 10 best-selling products (by quantity sold)**  
- **Sales distribution by payment method**  
- **Regional sales performance by customer location**

This analysis helps identify customer preferences, market trends, and sales strategies for business optimization.

```python
# Import package

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data 1

df1 = pd.read_excel(r"C:\Users\userr\OneDrive\Documents\Amazon Sales 2025 Project\Amazon 1.xlsx")
```

```python
df1.head(10)
```

```python
df1.isnull().sum()
```

```python
df1.info()
```

```python
# Total Sales by Product Category

plt.figure(figsize=(10,5))
sns.barplot(data=df1, x='Category', y='Total_Sales', palette='viridis')
plt.title("Total Sales by Category")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

This bar chart displays the total sales for each product category.

####  Key Insights
- **Electronics** is the top-performing category, generating more than **$55,000** in total sales — almost **twice** as much as any other category.
  
- **Home Appliances** comes in second, with sales close to **$30,000**, showing strong consumer demand in this segment.
- Categories like **Footwear**, **Clothing**, and especially **Books** contributed only a small fraction of the total revenue.
- The **Books** category had the **lowest sales**, indicating a potential area to either revamp or reduce inventory.

####  Business Takeaway:
Focus marketing and inventory investments on Electronics and Home Appliances, as they drive the majority of sales. Consider evaluating the viability of underperforming categories such as Books and Clothing.

```python
# Monthly sales trends

plt.figure(figsize=(10,5))
sns.lineplot(data=df2, x='Month', y='Monthly_Sales', marker='o')
plt.title("Total Monthly Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

This line chart shows the trend of total monthly sales from February to April 2025.

####  Key Insights
- **March 2025** recorded the **highest total sales**, slightly surpassing **$45,000**, showing strong performance during that month.
  
- **February 2025** also saw high sales, around **$41,000**, indicating consistent early-quarter demand.
  
- However, there was a **sharp decline in April 2025**, with total sales dropping drastically to under **$5,000**.

####  Business Takeaway:
The significant drop in April sales is a red flag and should be investigated further. Possible reasons could include stock shortages, seasonal effects, reduced marketing efforts, or external market changes. Proactive steps should be taken to identify the root cause and avoid prolonged sales decline in the upcoming months.

```python
# Top 10 best-selling products (by quantity sold)

plt.figure(figsize=(10,6))
sns.barplot(data=df3, x='Product', y='Total_Quantity', palette='magma')
plt.title("Top 10 Best-Selling Products")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- **Smartwatch** is the most sold product, slightly ahead of **Smartphone**, indicating strong demand for wearable tech.
- **Laptops** and **T-Shirts** follow, showing both tech and apparel have consistent interest from buyers.
- **Running Shoes** and **Jeans** suggest that activewear and casual fashion items are also popular.
- **Refrigerators** and **Washing Machines**, while less sold, still make the top 10—highlighting interest in home appliances.
- **Books** and **Headphones** round out the list, pointing to steady demand in both education and entertainment categories.

###  Business Takeaway

- **Tech products dominate** the top ranks, showing a potential area for increased marketing or bundling offers.
- **Fashion and apparel** (T-Shirts, Shoes, Jeans) also perform well—consider introducing new seasonal styles or cross-promotions.
- The presence of **home appliances** like refrigerators suggests potential for upselling or warranties.
- Lower sales for **Books** and **Headphones** could be explored further—maybe through targeted campaigns or product improvements.

```python
# Sales distribution by payment method

plt.figure(figsize=(10,6))
sns.barplot(data=df4, x='Payment_Method', y='Total_Sales', palette='coolwarm')
plt.title("Sales by Payment Method")
plt.tight_layout()
plt.show()
```

### Key Insights

- **PayPal** is the most preferred payment method, contributing the highest total sales by a large margin. This indicates that many customers trust and frequently use digital wallets.
- **Amazon Pay** and **Credit Cards** are also widely used, showing strong adoption of both branded digital payments and traditional card-based methods.
- **Debit Cards** have noticeably lower usage compared to credit-based options, possibly due to spending limits or user preference.
- **Gift Cards** account for the smallest portion of total sales, suggesting they are used occasionally, likely for specific promotions or as gifts.



### Business Takeaway

- **Enhancing PayPal promotions or simplifying the checkout** process for PayPal users could further boost conversions.
- Consider **bundling or incentives for Amazon Pay and Credit Card** users, as they represent a solid user base.
- Explore ways to **increase Gift Card engagement**, such as seasonal campaigns or loyalty rewards.
- The relatively low **Debit Card usage** might benefit from better education or confidence-building around its use online.

```python
# Regional sales performance by customer location

plt.figure(figsize=(10,6))
sns.barplot(data=df5, x='Customer_Location', y='Total_Sales', palette='Set2')
plt.title("Sales by Customer Location")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Regional Sales Performance: Customer Location Breakdown

- **Top Performing Cities**:
  - **Miami**, **Houston**, and **Los Angeles** lead in total sales, each generating over 11,000 in revenue.
  - These cities may represent strong market potential, possibly due to higher population density, spending power, or effective marketing reach.

- **Mid-Tier Markets**:
  - **Denver** and **Seattle** also show solid sales performance, suggesting consistent demand and possibly loyal customer bases in these regions.
  - These areas might be ideal for sustaining growth through retention strategies.

- **Lower Performing Cities**:
  - **Dallas**, **Boston**, and **Chicago** fall into the mid-to-lower range, indicating room for targeted campaigns or localized promotions to boost sales.
  - **San Francisco** and **New York** surprisingly rank at the bottom, which may be due to high competition, demographic mismatch, or under-penetration.



### Actionable Business Insights

- **Focus on High Performers**: Double down on Miami, Houston, and LA with exclusive deals, new product launches, or partnerships to maximize returns in high-conversion areas.
- **Reignite Lagging Regions**: Consider running localized ad campaigns or exploring customer feedback in New York and San Francisco to identify and address pain points.
- **Test Expansion in Mid-Tier Zones**: Cities like Denver and Seattle may respond well to loyalty programs or limited-time promotions to maintain momentum.



### Strategic Considerations

- Investigate **local economic factors** or **customer behavior trends** that contribute to the disparity.
- Ensure **inventory and delivery capabilities** match demand in high-performing regions.
- Use these insights to inform **geo-targeted marketing** and **sales resource allocation**.
